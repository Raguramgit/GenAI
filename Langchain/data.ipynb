```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.vectorstores import FAISS
loader = TextLoader("Unit_converter.txt")
documents = loader.load()
splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=100, chunk_overlap=5)
chunks = splitter.split_documents(documents)
embeded_new = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
embedding4 = embeded_new.embed_query(chunks)
print(embedding4[:3])
vectorstores = FAISS.from_texts(chunks,embedding4)
vectorstores.save_local("My 1st vectrodb")
query = "What is you company name"
result = vectorstores.similarity_search(query)
for r in result:
    print(r.page_content)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Load documents
loader = TextLoader("Unit_converter.txt")
documents = loader.load()

# Split into chunks (use the splitter variable consistently)
splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=100, chunk_overlap=5)
chunks = splitter.split_documents(documents)

# Create embeddings and build FAISS index from the chunked documents
embedder = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(chunks, embedder)
vectorstore.save_local("my_first_vectrodB")

# Example similarity search
query = "unit conversion from meters to kilometers"
results = vectorstore.similarity_search(query, k=3)
for r in results:
    print(r.page_content)
```
